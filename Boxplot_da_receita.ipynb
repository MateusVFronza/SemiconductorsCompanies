# Semiconductor companies, by revenue

## *Libraries*

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy
```

## *Read CSV*

```python
# The dataset contains commas and dots, we have to replace the dot and set the comma as decimal separator
    # Call pandas.to_numeric(arg, downcast=dtype) 
base = pd.read_csv('/content/Top semiconductor vendors, by revenue - Página5.csv', 
                   decimal=",")
base["Receita"]=base["Receita"].str.replace('.','')

df = pd.DataFrame(base)
df['Receita'] = pd.to_numeric(df['Receita'], downcast="float")
df
```

##### 

```python
df.describe().round(2)
```

```python
print('The median: ', df.Receita.median())
x = numpy.quantile(df.Receita, [0,0.25,0.5,0.75,1])
print('The quartiles: ',x)
```

```python
plt.rcParams['figure.figsize'] = (7,7)
plt.boxplot(df.Receita)
plt.show()
plt.savefig('boxplot_receita.png')
```
